Trong đó, El Niño - pha nóng của chu kỳ ENSO - là một trong những hình thái khí hậu cực đoan mang tính hủy diệt cao nhất, gây ra hạn hán nghiêm trọng tại các vựa nông sản lớn trên thế giới như Đông Nam Á và Nam Mỹ. ## **2. Phương pháp và Quy trình xử lý dữ liệu**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url_enso = (
    "https://drive.google.com/uc?id="
    "1FlHSjlOr8pHdZnCNncld_bSuyOI784Iu"
)

url_wb = (
    "https://drive.google.com/uc?id="
    "1lDcb6TLehXZmOAoJ-8I2eNl07pQAkO1o"
)

df_enso_raw = pd.read_csv(url_enso)
df_wb_raw = pd.read_csv(url_wb)

df_enso_raw.head()
# df_wb_raw.head()
```

### **2.1. Tiền xử lý và Hợp nhất dữ liệu**

```python
# Xoay trục dữ liệu kinh tế
df_wb_wide_raw = df_wb_raw.pivot_table(
    index='date',
    columns='commodity_name',
    values='price_nominal_usd'
).reset_index()

df_wb_wide_raw.rename(columns={
    'date': 'Date',
    'Rice, Thai 5%': 'Rice',
    'Coffee, Robusta': 'Robusta',
    'Coffee, Arabica': 'Arabica'
}, inplace=True)

df_enso = df_enso_raw[['Date', 'ONI']].copy()
df_wb_wide = df_wb_wide_raw[['Date', 'Rice', 'Robusta', 'Arabica']].copy()

df_enso['Date'] = pd.to_datetime(df_enso['Date'])
df_wb_wide['Date'] = pd.to_datetime(df_wb_wide['Date'])

df_merged = pd.merge(df_enso, df_wb_wide, on='Date', how='inner')

df = df_merged.dropna()
df.head()
```

### **Khử xu hướng dài hạn**
Phân tích tương quan ban đầu trên dữ liệu giá tuyệt đối thường gặp phải hiện tượng tương quan giả do các chuỗi số liệu đều chứa xu hướng tăng dài hạn theo thời gian. Để khắc phục hiện tượng này và đảm bảo tính dừng cho mô hình kinh tế lượng, nghiên cứu áp dụng kỹ thuật tính **Tốc độ tăng trưởng giá hàng tháng:**

Thay vì sử dụng mức giá tuyệt đối, toàn bộ dữ liệu giá hàng hóa được chuyển đổi thành tỷ lệ phần trăm thay đổi so với tháng trước thông qua hàm sai phân bậc nhất (`pct_change()`). Thao tác này giúp loại bỏ hoàn toàn yếu tố xu hướng dài hạn (bao gồm cả tác động tích lũy của lạm phát vĩ mô) và đưa dữ liệu về trạng thái dừng. Nhờ đó, các hệ số tương quan chéo (CCF) ở các bước sau sẽ phản ánh chính xác các cú sốc ngắn hạn do biến động khí hậu (El Niño) gây ra, thay vì bị nhiễu bởi đà tăng tự nhiên của thị trường.

```python
df_pct = df.copy()
df_pct['Rice_return'] = df_pct['Rice'].pct_change()
df_pct['Robusta_return'] = df_pct['Robusta'].pct_change()
df_pct['Arabica_return'] = df_pct['Arabica'].pct_change()

df_pct = df_pct.dropna()
df_pct.head()
```

#### **Trực quan hóa sự khác biệt giữa Giá gốc và Monthly Return**

Biểu đồ dưới đây sẽ minh họa rõ ràng cách chuyển đổi sang 'Monthly Return' loại bỏ xu hướng dài hạn và chỉ giữ lại các biến động tương đối của giá.

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

ax1.plot(df['Date'], df['Rice'], label='Giá Gạo Gốc (USD/tấn)', color='blue')
ax1.set_title('Giá Gạo Gốc theo thời gian', fontsize=14)
ax1.set_ylabel('Giá (USD/tấn)', fontsize=12)
ax1.legend()
ax1.grid(True, linestyle='--', alpha=0.7)

ax2.plot(df_pct['Date'], df_pct['Rice_return'], label='Monthly Return của Gạo', color='red')
ax2.set_title('Monthly Return của Gạo theo thời gian', fontsize=14)
ax2.set_xlabel('Ngày', fontsize=12)
ax2.set_ylabel('Monthly Return (%)', fontsize=12)
ax2.axhline(0, color='gray', linestyle='--')
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.suptitle('So sánh Giá Gạo Gốc và Monthly Return', y=1.02, fontsize=16, fontweight='bold')
plt.show()
```

### **2.2. Phân tích khám phá**

Trước khi đi sâu vào phân tích hệ số tương quan, tôi xin điểm qua Bảng thống kê mô tả của tập dữ liệu sau khi được xử lý.

```python
cols = ['ONI', 'Rice_return', 'Robusta_return', 'Arabica_return']

desc = df_pct[cols].describe().round(3)

desc.index = ['Số quan sát', 'Trung bình', 'Độ lệch chuẩn',
              'Min', 'Q1 (25%)', 'Trung vị (50%)', 'Q3 (75%)', 'Max']

desc.columns = ['Chỉ số ONI (El Niño)', 'Giá Gạo (Return)', 'Giá Cà phê Robusta (Return)', 'Giá Cà phê Arabica (Return)']
desc = desc.astype(object)
desc.loc['Số quan sát'] = desc.loc['Số quan sát'].astype(int).astype(str) + ' tháng'

desc
```

Bảng thống kê trên đây là kết quả của tập dữ liệu đã sạch, và nó xác nhận 3 điểm mấu chốt:

- **Quy mô dữ liệu:** Với count là 779 tháng, đây là một bộ dữ liệu đủ lớn, đảm bảo tính đại diện cao nhất cho các biến động thị trường trong hơn nửa thế kỷ qua.
- **Độ tinh khiết toán học:** Mọi người hãy chú ý hàng *mean* (giá trị trung bình). Tất cả đều lùi về mức xấp xỉ 0. Đây là 'chìa khóa' khẳng định chuỗi dữ liệu đã đạt trạng thái dừng hoàn hảo. Bằng cách chuyển sang tỷ lệ thay đổi hàng tháng, tôi đã triệt tiêu hoàn toàn đà tăng giá ảo do lạm phát dài hạn gây ra.
- **Bản chất kinh tế của nông sản:** Nhìn vào độ lệch chuẩn *(std)* và giá trị cực hạn, chúng ta thấy được sự phân hóa rất rõ:
Gạo (lương thực sinh tồn) có độ nhạy rất cao, dễ gây sốc khi đứt gãy cung cầu.
Cà phê thì có sự phân hóa đẳng cấp: Arabica (cao cấp, khó trồng) so với Robusta (đại trà, chịu hạn tốt).

Sự khác biệt này không chỉ là con số, nó phản ánh bản chất kinh tế và vị trí địa lý khác nhau. Chính đặc tính này sẽ dẫn đến 'phản ứng bất đối xứng' trước cùng một cú sốc khí hậu - và đó chính là điều tôi sẽ chứng minh bằng các con số định lượng ở phần tiếp theo."


**Giả thuyết nền tảng:** Tác động của sốc khí hậu lên giá nông sản **không** diễn ra tức thời vi tác động sinh học và chuỗi cung ứng luôn có độ trễ dẫn truyền.

Do đó, khi đo tương quan tại cùng một thời điểm
(Lag = 0), kết quả kỳ vọng phải **rất gần 0** - gần như không có mối liên hệ tức thời giữa ONI
và giá cả.

```python
cols = ['ONI', 'Rice_return', 'Robusta_return', 'Arabica_return']
corr_matrix = df_pct[cols].corr(method='pearson').round(3)

print(corr_matrix.to_string())
print()
r_rice    = df_pct['ONI'].corr(df_pct['Rice_return'])
r_robusta = df_pct['ONI'].corr(df_pct['Robusta_return'])
r_arabica = df_pct['ONI'].corr(df_pct['Arabica_return'])
print(f"Tương quan ONI <-> Giá Gạo: r = {r_rice:.3f}")
print(f"Tương quan ONI <-> Giá Cà phê Robusta: r = {r_robusta:.3f}")
print(f"Tương quan ONI <-> Giá Cà phê Arabica: r = {r_arabica:.3f}")

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True, fmt='.3f',
    cmap='coolwarm', vmin=-1, vmax=1,
    linewidths=0.5, linecolor='white',
    annot_kws={'size': 10, 'weight': 'bold'},
    square=True, ax=ax
)
ax.set_title('Heatmap Tương quan Tức thời (Lag = 0)\nONI vs Giá Gạo & Cà phê', fontsize=13, fontweight='bold', pad=15)
plt.tight_layout()
plt.savefig('heatmap_zero_lag.png', dpi=150, bbox_inches='tight')
plt.show()
```

Kết quả phân tích ma trận tương quan và biểu đồ nhiệt đã xác nhận rằng hệ số tương quan Pearson giữa Chỉ số ONI và giá của Gạo, Cà phê Robusta, Cà phê Arabica tại Lag = 0 đều nằm trong khoảng **[-0.03 ; 0.03]** - đây là một mức độ tương quan **rất yếu và không có ý nghĩa đáng kể về mặt thống kê.**

Điều này **chứng minh giả thuyết nghiên cứu là đúng**: Một chu kỳ El Niño bùng phát không gây ra cú sốc giá ngay lập tức trên thị trường nông sản. Tác động cần một khoảng thời gian trễ đủ dài để truyền dẫn qua toàn bộ chuỗi cung ứng thực tế phức tạp. Bước tiếp theo, chúng ta sẽ áp dụng kỹ thuật dịch chuyển chuỗi thời gian để xác định khung thời gian tối ưu mà tại đó mối tương quan trở nên mạnh mẽ và có ý nghĩa nhất.

### **2.3. Kỹ thuật Phân tích Độ trễ**
Đây là thuật toán cốt lõi của dự án. Áp dụng kỹ thuật **dịch chuyển chuỗi thời gian** kết hợp với cô lập sự kiện (El Niño).

Thay vì tính tương quan dàn trải, thuật toán hoạt động theo hai bước logic:
1. Dịch chuyển thời gian: Giữ nguyên trục thời gian của chỉ số ONI, nhưng tịnh tiến chuỗi Tỷ suất sinh lời của Gạo, Robusta và Arabica để đối chiếu ONI của tháng hiện tại với biến động giá của $k$ tháng sau đó (quét liên tục từ Lag = 0 đến Lag = 12 tháng).
2. Cô lập sự kiện: Tại mỗi bước tịnh tiến, máy tính chỉ trích xuất và tính toán hệ số tương quan Pearson trên những điểm dữ liệu mà El Niño thực sự bùng phát mạnh ($ONI \ge 0.5$).

**Mục tiêu:** Mốc thời gian (số tháng trễ) nào cho ra hệ số tương quan cao nhất, đó chính là "Độ trễ pha" tối ưu, phác họa chính xác khoảng thời gian vật lý cần thiết để một cú sốc khí hậu từ đại dương truyền dẫn qua chuỗi cung ứng và tác động mạnh nhất đến giá nông sản toàn cầu.

```python
lags = range(0, 13)
all_lag_correlations = []

for lag in lags:
    shifted_rice = df_pct['Rice_return'].shift(-lag)
    shifted_robusta = df_pct['Robusta_return'].shift(-lag)
    shifted_arabica = df_pct['Arabica_return'].shift(-lag)

    mask = df_pct['ONI'] >= 0.5

    corr_rice = df_pct.loc[mask, 'ONI'].corr(shifted_rice[mask])
    corr_robusta = df_pct.loc[mask, 'ONI'].corr(shifted_robusta[mask])
    corr_arabica = df_pct.loc[mask, 'ONI'].corr(shifted_arabica[mask])

    all_lag_correlations.append({
        'Độ trễ (tháng)': lag,
        'Hệ số Gạo': corr_rice,
        'Hệ số Robusta': corr_robusta,
        'Hệ số Arabica': corr_arabica
    })

df_lag_corr = pd.DataFrame(all_lag_correlations).round(3)
print("BẢNG TƯƠNG QUAN MỚI (Đã xử lý tính dừng & Lọc nhiễu El Nino):")
df_lag_corr
```

```python
idx_rice = df_lag_corr['Hệ số Gạo'].abs().idxmax()
idx_robusta = df_lag_corr['Hệ số Robusta'].abs().idxmax()
idx_arabica = df_lag_corr['Hệ số Arabica'].abs().idxmax()

best_lag_rice= df_lag_corr.loc[idx_rice, 'Độ trễ (tháng)']
best_r_rice = df_lag_corr.loc[idx_rice, 'Hệ số Gạo']

best_lag_robusta = df_lag_corr.loc[idx_robusta, 'Độ trễ (tháng)']
best_r_robusta = df_lag_corr.loc[idx_robusta, 'Hệ số Robusta']

best_lag_arabica= df_lag_corr.loc[idx_arabica, 'Độ trễ (tháng)']
best_r_arabica = df_lag_corr.loc[idx_arabica, 'Hệ số Arabica']

print("ĐỈNH TƯƠNG QUAN SAU KHI TỐI ƯU MÔ HÌNH (Dữ liệu đã khử xu hướng dài hạn & Lọc nhiễu):")
print("-" * 70)
print(f"Gạo     — Đỉnh tương quan: r = {best_r_rice:.3f} tại Lag = {best_lag_rice} tháng")
print(f"Robusta — Đỉnh tương quan: r = {best_r_robusta:.3f} tại Lag = {best_lag_robusta} tháng")
print(f"Arabica — Đỉnh tương quan: r = {best_r_arabica:.3f} tại Lag = {best_lag_arabica} tháng")
print("-" * 70)
```

Một sự kiện khí hậu như El Niño vừa khởi phát trên đại dương, nó không thể ngay lập tức gây ra thiệt hại mùa màng và ảnh hưởng đến nguồn cung trong cùng tháng đó, đặc biệt với cây dài ngày. - Sự thiếu hụt nguồn cung dẫn đến tăng giá trên thị trường quốc tế, thường được làm dịu bởi kho dự trữ của các tập đoàn và chính phủ (mất thêm thời gian).

**Kết quả thực tế từ dữ liệu:** Đối chiếu với ma trận tương quan ở Phần 2.2, giả thuyết trên về cơ bản là chính xác. Hệ số tương quan $r$ giữa Chỉ số ONI và các mặt hàng tại $Lag = 0$ đều ghi nhận ở mức cực kỳ thấp: Giá Gạo ($r = 0.011$), Robusta ($r = 0.027$) và Arabica ($r = -0.029$).

Các con số gần như bằng 0 này là khẳng định ban đầu: Một chu kỳ El Niño bùng phát không gây ra cú sốc giá ngay lập tức trên thị trường nông sản một cách đồng nhất, tạo tiền đề vững chắc để áp dụng kỹ thuật time-shifting ở các bước tiếp theo nhằm khám phá các độ trễ tiềm ẩn và sự phân hóa giữa các mặt hàng.

### **3.2. Xác định đỉnh tương quan và Độ trễ**

Kết quả từ kỹ thuật tịnh tiến chuỗi thời gian tại Phần 2.3 đã làm rõ mối quan hệ có độ trễ giữa El Niño và giá nông sản. Bảng xếp hạng tương quan cho thấy một sự phân hóa thú vị trong hành vi thị trường giữa các mặt hàng:

- **Gạo:** Đối với Gạo, đỉnh tương quan đạt **r = 0.115 tại Lag = 6 tháng**. Khác với các phản ứng hoảng loạn tức thời mang tính đầu cơ, con số Lag = 6 phản ánh độ trễ dự trữ chiến lược. Khi El Niño bùng phát, các quốc gia xuất khẩu và tiêu thụ lớn thường sử dụng kho dự trữ quốc gia để can thiệp và bình ổn giá trong những tháng đầu. Chỉ đến khoảng tháng thứ 6, khi kho dự trữ bắt đầu vơi đi và tác động của hạn hán bắt đầu phản ánh trực tiếp vào việc sụt giảm sản lượng thu hoạch thực tế, thị trường mới thực sự đối mặt với sự khan hiếm nguồn cung thực tế, đẩy giá lên thiết lập mặt bằng đỉnh mới.

- **Cà phê:** Với Cà phê Robusta, đỉnh tương quan đạt **r = 0.169 tại Lag = 9 tháng**. Với Arabica đỉnh tương quan đạt **r = 0.126 tại Lag = 9 tháng**. Sau khi loại bỏ nhiễu xu hướng dài hạn, dữ liệu cho thấy Arabica cũng đạt đỉnh độ trễ cùng thời điểm với Robusta (9 tháng), nhưng với cường độ tác động yếu hơn (0.126 < 0.169).

Các con số Lag = 6 và Lag = 9 là những minh chứng định lượng sắc bén, củng cố luận điểm cốt lõi của nghiên cứu: Chỉ số khí hậu đại dương không chỉ là dữ liệu thời tiết, mà chính là "chiếc la bàn" dẫn đường, cho phép chúng ta nhìn trước được quỹ đạo của nền kinh tế nông nghiệp toàn cầu từ nửa năm đến gần một năm.

### **3.3. Đối chiếu độ trễ với thực tế chuỗi cung ứng**

Các con số độ trễ tối ưu tìm được từ mô hình định lượng - bao gồm Lag = 6 tháng đối với Gạo và Lag = 9 tháng đối với Cà phê (Robusta & Arabica) - không phải là các kết quả ngẫu nhiên về mặt toán học. Chúng phản ánh một cách logic tiến trình truyền dẫn vật lý và kinh tế vĩ mô của một cú sốc khí hậu từ đại dương vào thị trường giao dịch hàng hóa thực tế.

Để kiểm chứng bằng kiến thức ngành, chuỗi nhân quả này có thể được bóc tách thông qua một mô hình truyền dẫn lý thuyết gồm 4 giai đoạn cốt lõi:

| Giai đoạn | Thời gian ước tính |
|---|---|
| El Niño bùng phát → Đảo lộn hoàn lưu khí quyển | 1-2 tháng |
| Hạn hán hình thành tại Đông Nam Á / Nam Mỹ | 1-2 tháng |
| Mất mùa, giảm sản lượng thu hoạch | 2-3 tháng |
| Báo cáo mùa vụ xác nhận thất thu → Cạn kiệt kho dự trữ vật chất → Giá lập đỉnh mới | 1-2 tháng |

*Lưu ý về tính linh hoạt của mô hình:* Khung thời gian cộng dồn từ 5 đến 9 tháng ở trên là một mô hình ước lượng mang tính bản chất. Trong thực tế, tốc độ truyền dẫn này không cố định mà sẽ dao động linh hoạt phụ thuộc vào:

(1) Biên độ và cường độ mạnh/yếu của từng đợt El Niño;

(2) Mức độ dày/mỏng của kho dự trữ quốc gia;

(3) Các chính sách can thiệp thị trường (như lệnh cấm xuất khẩu hoặc xả kho bình ổn giá) của các chính phủ.

Giải thích sự phân hóa thời gian giữa Gạo và Cà phê:
- **Gạo (Đỉnh trễ 6 tháng):**  Dữ liệu nghiên cứu sử dụng giá Gạo trắng Thái Lan 5% (giá FOB tại cảng Bangkok). Lúa là cây lương thực ngắn ngày với chu kỳ canh tác nhanh (3-4 tháng). Khi hạn hán hoặc mặn xâm nhập xảy ra, thiệt hại sinh học bộc lộ và được lượng hóa ngay tại thời điểm gặt. Tuy nhiên, thị trường cần thêm từ 2 đến 3 tháng sau thu hoạch để tiêu thụ nốt lượng kho dự trữ gối vụ cũ. Chỉ đến khi lượng gạo đổ về các kho cảng xuất khẩu thực sự sụt giảm nghiêm trọng, kết hợp với tâm lý nhạy cảm về an ninh lương thực kích hoạt làn sóng mua gom, giá FOB tại cảng Bangkok mới chính thức bị đẩy lên thiết lập mặt bằng đỉnh mới ở mốc 6 tháng.
- **Cà phê (Đỉnh trễ 9 tháng):** Dữ liệu nghiên cứu sử dụng giá chỉ báo quốc tế (ICO Indicator Prices) của Tổ chức Cà phê Quốc tế đối với hai dòng Robusta và Arabica. Cà phê là cây công nghiệp lâu năm có bộ rễ và sức chống chịu bền bỉ hơn cây lúa trong vài tháng đầu bằng cách tự sụt giảm dưỡng chất để nuôi cây. Tuy nhiên, khi chu kỳ thời tiết cực đoan (hạn hán đối với Robusta tại Đông Nam Á hoặc mưa bệnh đối với Arabica tại Nam Mỹ) dội thẳng vào giai đoạn thu hoạch chính diễn ra sau đó nhiều tháng, các báo cáo kiểm kê kho tồn kho thực tế tại các trung tâm tiêu thụ lớn (Âu - Mỹ) mới chính thức xác nhận sự khan hiếm nguồn cung đạt chuẩn, đẩy giá chỉ báo quốc tế lập đỉnh muộn hơn ở tháng thứ 9.


Đây là bằng chứng định lượng cho thấy dữ liệu khí hậu có thể đóng vai trò **tín hiệu dẫn
đường** sớm hơn 6 tháng so với diễn biến giá cả thực tế - nền tảng
trực tiếp cho Hệ thống Cảnh báo Sớm ở Phần 4.

### **3.4. Phân tích đối chiếu**

### **3.4.1. Phân tích tác động lên Giá Gạo: Hiện tượng "Phản ứng hai đỉnh"**

Biểu đồ phân tích dual-axis và ma trận độ trễ cho thấy giá Gạo không chỉ phản ứng một lần mà trải qua một chuỗi phản ứng hai giai đoạn cực kỳ đặc trưng mỗi khi "bức tường đỏ" El Niño (ONI ≥ 0.5) xuất hiện:
- **Đỉnh 1 - Cú sốc tâm lý và Đầu cơ (Lag = 1 tháng, r = 0.112)**: Ngay khi có tin bão khí hậu khởi phát, giá Gạo giật lên gần như tức thời. Đây chưa phải là tác động sinh học (vì lúa chưa chết trên đồng), mà hoàn toàn do tâm lý hoảng loạn. Các quốc gia xuất khẩu lớn thường lập tức ban hành lệnh hạn chế xuất khẩu để bảo vệ an ninh lương thực; trong khi giới thương lái ráo riết găm hàng.
- **Khoảng lặng giả tạo (Lag 2 đến Lag 5)**: Hệ số tương quan sụt giảm. Lúc này, các chính phủ bắt đầu xả kho dự trữ quốc gia để hạ nhiệt thị trường, bù đắp lượng cung thiếu hụt, tạo ra cảm giác thị trường đã bình ổn trở lại.
- **Đỉnh 2 - Cú sốc vật lý thực sự (Lag = 6 tháng, r = 0.115)**: Đây mới là đỉnh trễ khan hiếm nguồn cung thực tế và nguy hiểm nhất. Khi kho dự trữ dần cạn kiệt, cộng thêm đòn giáng từ hạn hán làm sụt giảm sản lượng thu hoạch thực tế, thị trường rơi vào trạng thái đứt gãy cung-cầu, đẩy mặt bằng giá thiết lập một đỉnh cao mới.

### **3.4.2. Phân tích đối chiếu Robusta và Arabica: Bất đối xứng về cường độ**

Đối với nhóm cà phê, biểu đồ cho thấy rõ ràng tác động bất đối xứng của El Niño lên hai loại hạt, mặc dù chúng có cùng độ trễ là 9 tháng. Sự khác biệt cốt lõi nằm ở cường độ tàn phá địa lý:
- **Đối với Robusta (r = 0.169)**: Các đợt El Niño luôn kéo theo những cú sốc giá bạo liệt. Nguyên nhân là Robusta tập trung trồng ở Đông Nam Á (Việt Nam, Indonesia) - nơi El Niño dội thẳng những đợt hạn hán khốc liệt nhất. Khi hoa cà phê không thể đậu quả do thiếu nước, sự đứt gãy nguồn cung là tuyệt đối và khó có thể đảo ngược, buộc giá bùng nổ mạnh mẽ.
- **Đối với Arabica (r = 0.126)**: Dù cũng đạt đỉnh tương quan ở tháng thứ 9, nhưng phản ứng của Arabica mờ nhạt và yếu hơn đáng kể. Arabica chủ yếu được trồng ở Nam Mỹ (Brazil, Colombia). Tại đây, hình thái El Niño phức tạp và phân tán hơn: nó có thể gây khô hạn ở phía Bắc Brazil nhưng lại mang mưa lớn tới khu vực phía Nam. Sự phân tán rủi ro về mặt địa lý này giúp tổng nguồn cung Arabica toàn cầu không bị sụp đổ toàn diện như Robusta.

### **Tiểu kết:**
Sự phân hóa trong hành vi giá này nhấn mạnh rằng El Niño không phải là một "lưỡi hái" cắt đều lên mọi loại nông sản. Việc nhận diện được "khoảng lặng giả tạo" của Gạo và "tác động bất đối xứng" của Cà phê là chìa khóa sống còn để thiết kế các chiến lược phòng ngừa rủi ro ở Phần 4.

### **4. Đề xuất Giải pháp - Mô hình Cảnh báo Sớm**

### **4.1. Chiến lược bình ổn giá của Chính phủ: 'Khung thời gian vàng' cho Gạo**

Dựa trên hiện tượng "Phản ứng hai đỉnh" của giá Gạo, Chính phủ cần thực hiện chiến lược can thiệp phân kỳ để bảo toàn nguồn lực quốc gia:
- **Tháng thứ 1 - 2**: Ngay khi ONI $\ge$ 0.5, ưu tiên dùng các công cụ hành chính (rà soát chống đầu cơ, kiểm soát hạn ngạch xuất khẩu) để trấn an thị trường. Tránh xả kho dự trữ sớm vì thiệt hại sản lượng thực tế chưa diễn ra.
- **Tháng thứ 3 - 5 (Giai đoạn Dự phòng)**: Tận dụng "khoảng lặng" để gia cố kho dự trữ quốc gia. Đồng thời triển khai khẩn cấp các biện pháp kỹ thuật (điều tiết thủy lợi, chuyển đổi giống chịu hạn) nhằm giảm thiểu thiệt hại sinh học cho vụ mùa.
- **Tháng thứ 6 trở đi (Giai đoạn Vật lý)**: Đây là "điểm rơi" của sự khan hiếm nguồn cung thực tế. Chỉ triển khai xả kho dự trữ vào thời điểm này để bù đắp sản lượng thiếu hụt, giúp triệt tiêu đà tăng giá và bảo đảm an ninh lương thực.

**Tóm lại:** Không xả kho quá sớm ở tháng thứ 1 (khi chỉ có tin đồn), mà phải giữ lực để can thiệp đúng tháng thứ 6 (khi sản lượng thực tế sụt giảm).

### **4.2. Chiến lược phòng ngừa rủi ro cho Doanh nghiệp: Tối ưu hóa chuỗi cung ứng Cà phê**

Với mốc Lag = 9 tháng và tính chất bất đối xứng cường độ ($r_{\text{Robusta}} = 0.169 > r_{\text{Arabica}} = 0.126$), các tập đoàn F&B toàn cầu như Starbucks, Nestlé, El Niño không chỉ là rủi ro mà còn là cơ hội để tối ưu hóa chuỗi cung ứng nếu có đủ thông tin và chiến lược phù hợp:
- **Phòng vệ giá bằng Hợp đồng Tương lai:** Ngay khi hệ thống phát tín hiệu ONI $\ge$ 0.5, doanh nghiệp có một "cửa sổ cơ hội" kéo dài tối đa 9 tháng trước khi giá thực tế lập đỉnh. Cần lập tức mua trước các hợp đồng tương lai để "khóa" giá Robusta trên sàn London, triệt tiêu hoàn toàn rủi ro tăng giá đỉnh điểm của nguồn cung Đông Nam Á ở tháng thứ 9.
- **Linh hoạt cấu trúc pha trộn:** Vì Arabica có tính phân tán rủi ro địa lý tốt hơn tại Nam Mỹ nên chịu tác động yếu hơn Robusta ($0.126 < 0.169$). Doanh nghiệp cần chuẩn bị sẵn sàng phương án kỹ thuật: Thay đổi linh hoạt tỷ lệ phối trộn (Blend recipe) bằng cách tăng tỷ trọng Arabica và giảm bớt Robusta trong giai đoạn cao điểm của bão khí hậu để bảo vệ biên lợi nhuận.

**Tóm lại:** Biến dữ liệu khí hậu thành lợi thế cạnh tranh. Khung thời gian trễ 9 tháng là khoảng đệm lý tưởng để doanh nghiệp vừa kịp chốt hợp đồng giá tốt, vừa kịp điều chỉnh công thức sản phẩm trước khi thị trường toàn cầu rơi vào khủng hoảng nguồn cung.

### **4.3. Mô phỏng Hệ thống Cảnh báo Sớm bằng Python**

```python
def early_warning_system(oni_index, lag_rice=best_lag_rice, lag_robusta=best_lag_robusta, lag_arabica=best_lag_arabica):
    print(f"--- HỆ THỐNG PHÂN TÍCH TÍN HIỆU ONI: {oni_index} ---")

    if oni_index < 0.5:
        print("\033[92m[TRẠNG THÁI XANH] An toàn. Nhiệt độ đại dương bình thường, nguồn cung ổn định.\033[0m\n")

    elif 0.5 <= oni_index < 1.5:
        print("\033[93m[CẢNH BÁO VÀNG] Có dấu hiệu El Niño khởi phát (Mức độ nhẹ/trung bình).\033[0m")
        print("  • Chính phủ: Bắt đầu rà soát kho dự trữ, siết chặt quản lý xuất khẩu.")
        print("  • Doanh nghiệp: Đưa vào tầm ngắm các hợp đồng tương lai phòng vệ rủi ro.\033[0m\n")

    else:
        print("\033[91m[CẢNH BÁO ĐỎ] EL NIÑO CỰC ĐOAN! RỦI RO ĐỨT GÃY CHUỖI CUNG ỨNG NGHIÊM TRỌNG.\033[0m")

        print("\033[91m[!] DỰ BÁO QUỸ ĐẠO GIÁ GẠO:\033[0m")
        print("  • Đỉnh 1 (Tâm lý & Đầu cơ): Sẽ giật tăng ngay trong 1-2 tháng tới.")
        print(f"  • Đỉnh 2 (Khan hiếm vật lý): Dự kiến thiết lập mặt bằng đỉnh mới sau {lag_rice} tháng.")

        print("\033[91m[!] DỰ BÁO QUỸ ĐẠO GIÁ CÀ PHÊ:\033[0m")
        print(f"  • Robusta: Nguy cơ hạn hán Đông Nam Á khốc liệt, giá lập đỉnh sau {lag_robusta} tháng.")
        print(f"  • Arabica: Phản ứng mờ nhạt hơn, đạt đỉnh độ trễ sau {lag_arabica} tháng.")

        print("\033[91m[!] KHUYẾN NGHỊ CHIẾN LƯỢC CẤP BÁCH:\033[0m")
        print(f"  • Chính phủ: Giữ vững kho dự trữ, chỉ bung xả vào đúng tháng thứ {lag_rice}.")
        print("  • Doanh nghiệp: Mua Hợp đồng Tương lai Robusta ngay, linh hoạt tăng tỷ lệ Blend Arabica.\n")

# Kiểm thử hệ thống
early_warning_system(1.8)
early_warning_system(1.0)
early_warning_system(0.3)
```
